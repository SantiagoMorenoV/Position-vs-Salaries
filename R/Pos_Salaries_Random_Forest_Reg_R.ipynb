# **Did the interviewee tell the truth?**

An HR recruiter wants to know if a candidate told the truth or not about his salary in his previous position.

For this case, I have a dataset with the positions and salaries from the company where the candidate used to work.

# **Importing the dataset**

````r
dataset = read.csv('https://raw.githubusercontent.com/SantiagoMorenoV/Position-vs-Salaries/main/Position_Salaries.csv')
dataset
dataset = dataset[2:3]
dataset
````

With dataset = dataset[2:3], I keep the second and third columns, which are the columns that I will use in the model.

# **Plotting Data**

There are two ways of plotting Salary vs Position. With the library ggplot and by using the R's function "plot()" to create a scatter plot.

````r
library(ggplot2)
library(scales)
ggplot(data = dataset, aes(x = Level, y = Salary)) +
  geom_point(color = "red") +
  labs(x = "Position", y = "Salary", title = "Position vs Salary") +
  theme(plot.title = element_text(size = 16, face = "bold", hjust = 0.5),
        axis.text.x = element_text(size = 12),
        axis.text.y = element_text(size = 12)) +
  scale_y_continuous(labels = comma)
````

````r
plot(dataset$Level, dataset$Salary, col = "red", pch = 19,
     xlab = "Position", ylab = "Salary", main = "Position vs Salary")
````

# **Descriptive Statistics**

````r
summary(dataset)
````

# **Salary distribution**

````r
hist(dataset$Salary, main ="Salary's Histogram", col= '#0E4C92')
````

# **Boxplot**

````r
#install.packages("ggplot2")
library(ggplot2)
library(scales)
variables <- c("Salary")
for (var in variables) {
  print(ggplot(data = dataset, aes(x = "", y = dataset[,var])) +
    geom_boxplot(fill = "#0E4C92") +
    labs(x = "", y = var, title = paste0(var, "'s Boxplot")) +
    scale_y_continuous(labels = comma, limits = c(0, max(dataset[,var]))) +
    theme_bw() +
    theme(plot.title = element_text(hjust = 0.5, size = 16),
          axis.text.y = element_text(size = 12)))
}
````

# **Fitting the Random Forest Regression on the whole dataset**

````r
#install.packages('randomForest')
library(randomForest)
set.seed(1234)
regressor = randomForest(x = dataset[-2],
                         y = dataset$Salary,
                         ntree = 500)
print(regressor)
````

````r
# install.packages("caret")
library(caret)

# Predict salary values using the random forest model
predicted_salary <- predict(regressor, data.frame(Level = dataset$Level))

# Calculate R-squared value
rsq <- summary(lm(predicted_salary ~ dataset$Salary))$r.squared

# Print R-squared value
print(paste0("R-squared: ", round(rsq, 4)))
````

# **Predicting the salary**

````r
y_pred = predict(regressor, data.frame(Level = 6.5))
round(y_pred,2)
````

# **Visualising the Decision Tree Regression results (for higher resolution and smoother curve)**

````r
# install.packages('ggplot2')
library(ggplot2)
x_grid = seq(min(dataset$Level), max(dataset$Level), 0.01)
ggplot() +
  geom_point(aes(x = dataset$Level, y = dataset$Salary),
             colour = 'red') +
  geom_line(aes(x = x_grid, y = predict(regressor, newdata = data.frame(Level = x_grid))),
            colour = 'blue') +
  ggtitle('Truth or Bluff (Random Forest Regression)') +
  xlab('Level') +
  ylab('Salary')
````

**Note on the library used for obtaining the R-squared**

I did not implement the "rsq" library like in other models of this very repository, where the determination coefficient was easily obtained with the code:

```
summary(regressor)$r.squared
rsq.rpart(regressor)
```
Instead, I have utilized the "caret" library.

The `summary(regressor)$r.squared` method only provides the R-squared value for the linear regression case, and may not work for other models such as Random Forest Regressions. Similarly, `rsq.rpart(regressor)` is only applicable for regression trees. In contrast, the "caret" package provides a more comprehensive and flexible approach to model training, tuning, and evaluation, allowing for the calculation of a wide range of evaluation metrics including R-squared, regardless of the type of regression model used.
